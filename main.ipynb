```python
import numpy as np
import matplotlib.pyplot as plt

X_data = np.random.random(50) * 100 # Generate values between 0 and 1
y_data = np.random.random(50) * 100
```

# Scatter plot

```python
plt.scatter(X_data, y_data)
plt.show()
```

# Line plots

```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83, 80,
           82, 82, 83, 81, 80, 79]

# print(len(years))
# print(len(weights))

plt.plot(years, weights)
plt.show()
```

# Bar Plots

```python
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 1, 45]

plt.bar(x, y)
plt.show()
```

# Histograms

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages)
plt.show()
```

# Pie charts

```python
langs = ["C++", "C#", "Python", "Java", "Go"]
votes = [25, 20, 40, 5, 10]

plt.pie(votes, labels=langs, autopct="%.2f%%")
plt.show()
```

# Boxplot

```python
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
plt.show()
```

# Plot customization

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55, 56, 62, 61, 72, 72, 73, 75]
income_ticks = list(range(50, 81, 2)) # for the y axis

plt.plot(years, income)
plt.title("Income of John ($)", fontsize=20, fontname="FreeSerif")
plt.xlabel("Year")
plt.ylabel("Yearly income in $US")
plt.yticks(income_ticks, [f"{x}k$" for x in income_ticks])
plt.show()
```

# Legends and multiple plots

```python
stock_a = [100, 102, 99, 101, 101, 100, 102]
stock_b = [90, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]

plt.plot(stock_a, label="Company 1")
plt.plot(stock_b, label="Company 2")
plt.plot(stock_c, label="Company 3")
plt.legend(loc="lower center") # Without this command, doesn't work
plt.show()
```

```python
votes = [10, 2, 5, 16, 22]
people = ["A", "B", "C", "D", "E"]
plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

# Plot styling

```python
from matplotlib import style

style.use("dark_background")
votes = [10, 2, 5, 16, 22]
people = ["A", "B", "C", "D", "E"]
plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

# Multiple figures

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

style.use("default")

plt.figure(1)
plt.scatter(x1, y1)
plt.figure(2)
plt.plot(x2, y2)
plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Logarithm")

axs[1, 1].set_xlabel("Number")

fig.suptitle("Four plots")

plt.savefig("fourplots.png")

plt.show()
```

# 3D Plotting

```python
ax = plt.axes(projection="3d")
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3d plot")
ax.set_xlabel("Test")
plt.show()
```

```python
ax = plt.axes(projection="3d")
x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title("3d plot")
ax.set_xlabel("Test")
plt.show()
```

```python
ax = plt.axes(projection="3d")
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap="Spectral")
plt.show()
```
